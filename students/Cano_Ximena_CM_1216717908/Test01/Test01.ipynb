**2. Rutina CoolingFunction, mediante interpolación lineal. **

```python
def CoolingFunction(LT,M):
    #Obtenemos los datos de la Metalicidad.
    X=M[:,0]
    Y=M[:,5]
    
    try:
        n=len(LT)
    except:
        n= 1
        LT=np.array([LT,])
    
    term=lambda LT,i:(Y[i+1]-Y[i])/(X[i+1]-X[i])*(LT-X[i])+Y[i]

    #Entontremos el intervalo.
    yy=[]
    m=len(X)
    for n in xrange(0,n):
        for i in xrange(0,m):
            if LT[n] <= X[i]:
                break
        yy+=[term(LT[n],i-1)]
    yy=np.array(yy)
    return yy;
```

**3. Función Temperatura.**

```python
def Temperature(LT,M,t,nm):
    D=CoolingFunction(LT,M)
    g=lambda t,i:D-t
    
    #Método de punto fijo 
    pi=[t,]
    for i in xrange(0,nm+3):
        tt=g(pi[i],i)
        pi+=[tt]
        t=tt
    return pi[-1];
```
