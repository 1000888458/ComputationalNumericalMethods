**1. El formalismo Press-Schechter:** El siguiente scrit the python hace:
* Carga el catálogo de los halos de materia oscura.
* Crea una grafica de un histograma de log10 (masa) de los halos.
* Crea, en la misma figura, una grafica de la predicción teórica del formalismo de Press-Schechter.

```python
#! usr/bin/python
#-*- coding:utf-8 -*-
%pylab inline
import numpy as np
import matplotlib.pyplot as mt

logs=[]
#Cargamos los datos
halos=np.loadtxt("FOF.dat")
#Cuardamos en listas cada columna del catálogo 
x=halos[:,0]
y=halos[:,1]
z=halos[:,2]
M=halos[:,3]
Ms=np.array([M])
ms=[]
def N(masa):
    m=1.08e14*M**(-0.889)*np.exp(-9.7e-32*M**0.222)
    return m
#Creamos la figura
mt.subplot(1,2,1)
mt.hist(np.log10(M),bins=200,color="purple")
mt.title("Histograma")
mt.grid()
mt.xlabel("$Log10(masa)$")
mt.subplot(1,2,2)
mt.grid()
mt.plot(np.log10(M),N(M),color="purple")
show("Histogram.png")
```

**2. Distribución de Halos Masivos: **El siguiente script de python hace:

* Selecciona y grafica los halos más masivos.
* Grafica en 3D el catálogo completo de los Halos.

```python
#Importamos la librería de Axes para graficar en 3D
from mpl_toolkits.mplot3d import Axes3D

#Encontramos los halos más masivos
P=M>1*10**13

#x,y,z en función de P
Xm=x[P]
Ym=y[P]
Zm=z[P]

#Creamos la figura
figura=mt.figure(figsize=(8,8))
ax=figura.add_subplot(111,projection='3d')
ax.set_xlim(0,250)
ax.set_ylim(0,250)
ax.set_zlim(0,250)
ax.scatter(Xm,Ym,Zm,alpha=0.4,s=5,color="purple")
show("grafico3D.png")
figura2=mt.figure(figsize=(8,8))
ax=figura2.add_subplot(111,projection='3d')
ax.set_xlim(0,250)
ax.set_ylim(0,250)
ax.set_zlim(0,250)
ax.scatter(x,y,z,alpha=0.2,s=0.01,color="purple")
show("grafico3D2.png")
```
