## Representación Binaria   

**Precisión simple 32-bits.**

Un número en coma flotante de precisión simple es almacenado en 32 bits. Los bits se establecen de la siguiente manera: 

*|_S(1 bit)_|______Exponente(8 bits)______|________Mantisa(23 bits)________|*
                  

**Obtener la representación binaria en Float32 de un número real.**

Primero debemos encontrar el bit para el signo. Si el número es positivo se asigna un 0 (cero) y si el número es negativo se asigna un 1. 

Luego, escribimos la representación binaria del valor absoluto del número. 
Una manera simple de obtener la representación binaria de un número es con divisiones sucesivas entre 2, y leyendo el residuo desde el último hasta el primero.

Ahora, si el número es decimal, obtenemos la representación binaria de su parte entera. Posteriormente, Obtenemos la representación binaria de su parte fraccionaria. Esto se hace multiplicando sucesivamente por 2, si el resultado de la multiplicación es mayor o igual a 1, tomamos su parte entera (1) y si es resultado es menor a 1 o igual a 0, entonces tomamos su parte entera (0). esta operación se da por terminada cuando el resultado de la multiplicación es igual a 1 o 0.

Ahora, el exponente se obtiene normalizando el binario y sumando 127 al exponente de la normalización. Luego, encontramos su binario. 

En la mantisa, suprimimos el primer 1. Si la mantisa no cuenta con 23 bits, lo restante secompleta con ceros, y si excede los 23 bits, se suprime la cantidad necesaria para quedar de exactamente 23 bits.

**Ejemplo**

*Encontremos la representación binaria de 147.5

Vemos que el número es positivo, por lo cual, el bit para el signo es 0.

Por otra parte, la representación en binario de su parte entera (147) está dado por:

147/2=73 Residuo 1

73/2=36 Residuo 1

36/2=18 Residuo 0

18/2=9 Residuo 0

9/2=4 Residuo 1

4/2=2 Residuo 0

2/2=1 Residuo 0

Así, 147 es 10010011

**Nota:** Sí el ultimo resultado de la divición es 1, entonces, tomamos este como el primer bit del binario.

Además, encontramos el binario de 0.5, el cual está dado por:

0.5*2=1 

Así su binario es 1. 

Por lo cual tendríamos el total del binario como 10010011.1 

El exponente lo hallamos corriendo el punto hasta el primer 1. En este caso sería 1.00100111x10^7.

Tomamos 7 y sumamos 127, obteniendo 134, y su binario es 10000110

Por ultimo, eliminamos el 1 anterior al punto del binario de la mantisa normalizada, y unimos signo, exponente y mantisa para así obtener la representación binaria. Como la mantisa tiene un número menor a 23 bits, lo restante es completado con ceros. 

La representación float32 de 147.5 es:

01000011000100111000000000000000


### El siguiente script encuentra la representación en float 32 de un número real.

```python
#! usr/bin/python
# -*- coding:utf-8 -*-

#Rutina para hallar el binario de números enteros 
num=[]
def binario(nn):  
	ms=[]
        while nn>=2:  
		if nn==2:	
			ms+=[0]
		else:
			m=nn%2
               		ms+=[m]
		nn=nn/2
		if nn<2:
			ms+=[nn]
        ms=ms[::-1]
        return ms 
#Rutina para el binario de la parte decimal de un número fraccionario
def bindecimal(de):
	ts=[]
	if de==0:
		return ts
	else:
		while de<2:
			t=de*2
			tt=int(t)
			ts+=[tt]
			if t>0:
				de=t-tt
			else:
				de=t
			if de==0:
				de=2
		return ts 		
n=input("Ingrese numero: ")
# Obtenemos el signo 
s=[]
if n<0: 
	s+=[1]
	n=-1*n
else:
	s+=[0]
#Obtenemos la parte entera
nn=int(n)
#Obtenemos la parte decimal 
de=n-nn
#Binario del entero  
entero=binario(nn)
#Binario del decimal 
decimal=bindecimal(de)
#Mantisa
x=entero+decimal 
mantisa=x[1:]
#Encontramos el exponente y su binario 
a=len(entero)-1
exp=a+127
exponente=binario(exp)
#Obtenemos el total de 23 bits para la parte entera 
if len(mantisa)>23:
	mantisa=mantisa[:22]
while len(mantisa)<23:
	mantisa+=[0]
#Representación del número como float32
float32=s+exponente+mantisa

print "Representacion float32 de",n,":",float32
```
