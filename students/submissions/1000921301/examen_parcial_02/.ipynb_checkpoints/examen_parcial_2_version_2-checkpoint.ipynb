Para un dispositivo óptico particular que consiste en un medio no lineal dentro de una cavidad óptica, esto sólo ocurre cuando la intensidad I se encuentra dentro de un cierto rango que está dado por la ecuación: 

$$f(x)=(x+1)^2 - c(x-1) < 0 $$

donde $x$ es la intensidad del campo eléctrico y $c$ es la fuerza de la absorción del 
medio no lineal. 

## 1. 
Cree una función llamada `f_cost` que reciba dos parámetros (x,c) y solo devuelva la función f(x) como se mencionó anteriormente.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as optimize
```

```python
# tu codigo:
def f_cost(x, c):
  return (x+1)**2 - c*(x-1)
```

## 2. 
Usando funcion `f_cost` haz un gráfico con cuatro curvas para c=5,10,15,20 y con x entre -10,20 (50 puntos)
* hacer que cada curva sea de un color diferente
* añadir etiquetas

```python
# tu codigo:
C = [5, 10, 15, 20]
X = np.linspace(-10, 20, 50)

fig, ax = plt.subplots(figsize = (10,7))
plt.style.use("seaborn")
ax.plot(X, f_cost(X, C[0]), color = "darkblue", label = f"c = {C[0]}")
ax.plot(X, f_cost(X, C[1]), color = "darkgreen", label = f"c = {C[1]}")
ax.plot(X, f_cost(X, C[2]), color = "darkred", label = f"c = {C[2]}")
ax.plot(X, f_cost(X, C[3]), color = "purple", label = f"c = {C[3]}")
ax.axhline(y=0, xmin=0.05, xmax=0.95, color = "gray", linestyle = 'dashed')
plt.axvline(x=0, ymin=0.05, ymax=0.95, color = "gray", linestyle = 'dashed')
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend()
print("")
```

## 2a. Para cual `c` esta funcion no tiene raices?

```python
# tu respuesta
# Graficamente vemos que para c = 5 la función no tiene raíces, vamos a verificar
# esto usando scipy.optimize.
def root_newton(f_cost, start, c): 
    try:
        root = optimize.newton(f_cost, start, args=(c,))
    except:
        root=np.nan
    return root

r1 = root_newton(f_cost, 4, 5)
r1
```

## 3. 
Cree una función llamada `root_finding()` que devuelva el valor de la raíz (si existiera) para cada parametro `c` (pense `Newton` o `bisecion`). Usando esta funcion vamos a buscar primero raiz izquierda y despues raiz derecha.

```python
# tu codigo:
def root_finding(f_cost, start, c): 
    try:
        root = optimize.newton(f_cost, start, args=(c,))
    except:
        root=np.nan
    return root

for c in C:
  print(f"Para c = {c}, raiz izquierda = {root_finding(f_cost, -1, c):.2f} y raíz derecha = {root_finding(f_cost, 10, c):.2f}")
  
```

## 4a. 
Primero vamos a mirar el comportamiento de raiz de lado izquierda: Defina el vector `c_vec` que contiene 50 valores de 5 a 100. En el loop, llame a la función `root_finding()` para cada valor `c` en el vector `c_vec`. Guarda los resultados en variable `left_roots`

```python
# tu codigo:
c_vec = np.linspace(5, 100, 50)

left_roots = []
for c in c_vec:
  left_roots.append(root_finding(f_cost, -1, c))
```

## 4b 
Hacer un gráfico de raíz izquierda vs c

```python
# tu codigo:
fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(c_vec, left_roots, color = "darkred", label = "$Left$ $root$")
ax.set_title("f_cost left roots",
             fontdict={'family': 'serif', 
             'color' : 'k',
             'weight': 'bold',
             'size': 18})
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend()
print("")

# Tiene un comportamiento exponencial decreciente.
```

## 5a. 
Ahora vamos a mirar el comportamiento de raiz de lado derecha: Defina el vector `c_vec` que contiene 50 valores de 5 a 100. En el loop, llame a la función `root_finding()` para cada valor `c` en el vector `c_vec`. Guarda los resultados en variable `right_roots`.

```python
# tu codigo:
c_vec = np.linspace(5, 100, 50)

right_roots = []
for c in c_vec:
  right_roots.append(root_finding(f_cost, 50, c))
```

## 5b
* Hacer un gráfico de raíz derecha vs c

```python
# tu codigo:
fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(c_vec, right_roots, color = "darkred", label = "$right$ $root$")
ax.set_title("f_cost right roots",
             fontdict={'family': 'serif', 
             'color' : 'k',
             'weight': 'bold',
             'size': 18})
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend()
print("")

# Tiene un comportamiento lineal creciente.
```

## 6. 
Pregunta:
* este cierto *rango* para cual se obtiene un comportamiento complejo esta convergente o divergente cuando `c` es creciente
* si este rango esta convergiendo, ¿a qué valor?
* si este rango esta divergente, como esta divergente (lineal, cuadrático, diferente)

```python
# respuesta:
# Para las raíces izquierdas podemos ver que su valor a medida que c crece está 
# convergiendo, ya que decrece exponencialmente, hacía un valor de 1.

# Para las raíces derechas se puede que su valor a medida que c crece está 
# divergiendo, ya que crece infinitamente de forma lineal.

# Como las raíces izquierdas convergen a 1, no afecta mucho al comportamiento del
# rango entre las raíces a medida que c crece, el comportamiento de este rango
# será casi totalmente determinado por el comportamiento de las raíces derechas
# las cuales divergen linealmente, por lo tanto, el rango entre las raíces será 
# divergente y de una forma lineal.
```
