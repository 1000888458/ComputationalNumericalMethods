```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/content/sample_data/mnist_test.csv")
```

```python
ran = np.arange(10000)
ranl = np.random.choice(ran, 25)
L = []
t = []
for i in ranl:
  img = df.iloc[i].to_numpy()
  L.append(np.reshape(img[1:], (28, 28)))
  t.append(img[0])

fig, ax = plt.subplots(5, 5, constrained_layout=True, figsize=(10,7))

cont = 0
for i in range(5):
  for j in range(5):
    ax[i, j].imshow(L[cont])
    ax[i, j].set_title(t[cont])
    cont = cont + 1
```

* pista: comienza con digits.keys()

```python
from sklearn import datasets
digits = datasets.load_digits() # digits is a dictionary
```

```python
images = digits["data"]
rand = np.arange(len(images))
randl = np.random.choice(rand, 25)

imgL = []
tt = []
for i in randl:
  a = np.reshape(images[i], (8, 8))
  imgL.append(a)
  tt.append(digits["target"][i])
fig, ax = plt.subplots(5, 5, constrained_layout=True, figsize=(10,7))

count = 0
for i in range(5):
  for j in range(5):
    ax[i, j].imshow(imgL[count])
    ax[i, j].set_title(tt[count])
    count = count + 1
```

```python
import tensorflow as tf
(X_data, y_data), (_, _) = tf.keras.datasets.cifar10.load_data()
# X_data (numpy array) contiene los imagenes 
# y_data (numpy array) contiene las etiquietas
```

```python
random = np.arange(len(X_data))
randoml = np.random.choice(random, 25)

imageL = []
title = []
for i in randoml:
  imageL.append(X_data[i])
  title.append(y_data[i])

fig, ax = plt.subplots(5, 5, constrained_layout=True, figsize=(10,7))

contdr = 0
for i in range(5):
  for j in range(5):
    ax[i, j].imshow(imageL[contdr])
    ax[i, j].set_title(title[contdr])
    contdr = contdr + 1
```

```python
# Anteriores imagenes pero en escala de grises.
fig, ax = plt.subplots(5, 5, constrained_layout=True, figsize=(10,7))

contdr = 0
for i in range(5):
  for j in range(5):
    img_avg = np.average(imageL[contdr], -1)
    ax[i, j].imshow(img_avg)
    ax[i, j].set_title(title[contdr])
    contdr = contdr + 1
```
