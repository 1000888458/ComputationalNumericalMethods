```python
import numpy as np
import matplotlib.pyplot as plt

def parabola(x):
  return x**2 - 11

def parabola_prima(x):
  return 2*x

def newton_raphson(p, tol, f, f_prime): # p: Punto inicial, tol: Tolerancia
  L = []                                # f: FunciÃ³n, f_prime: Derivada de f.
  val = abs(f(p))
  while val > tol:
    xn = p - (f(p) / f_prime(p))
    L.append(xn)
    p = xn
    val = abs(f(p))
  return L

yl = newton_raphson(1, 0.0001, parabola, parabola_prima)
xl = np.linspace(0, len(yl), len(yl))

plt.style.use("seaborn")
fig, ax =  plt.subplots()

ax.plot(xl, yl, label = f"$Valor$ $final = {yl[-1]:.3f}$", color = "darkred")
plt.title("$Newton-Raphson$ $(F(x) = x^2 - 11,$ $tol = 0.0001)$")
ax.set_xlabel("$IteraciÃ³n$")
ax.set_ylabel("$Valor$ $de$ $la$ $raÃ­z$")
ax.legend()

print(f"NÃºmero de iteraciones: {len(yl)}")
```
