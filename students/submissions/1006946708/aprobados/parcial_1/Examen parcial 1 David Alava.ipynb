```python
 import math

import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=0.2
cos_x=0
for i in range (0,3):
    numerador=((-1)**i)*(x**(2*i))
    denominador=math.factorial(2*i)
    razon=numerador/denominador
    cos_x=cos_x + razon
print(cos_x)    
```

```python
def cos_taylor(x,k):
    cos_x=0
    
    for k in range (0,k+1):
        numerador=((-1)**k)*(x**(2*k))
        denominador=math.factorial(2*k)
        razon=numerador/denominador
        cos_x=cos_x + razon
    return cos_x
print(cos_taylor(0.2,2))
    
```

```python
xvec=np.linspace(-np.pi,np.pi,100)
#xvec=np.arange(0,100)
cos_aprox=[]
for i in range(0,100):
   
    cos_aprox.append(cos_taylor(xvec[i],3))
print(cos_aprox)   
    
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,5))
ax.plot(xvec,cos_aprox,'--r',label='taylor cos(x)')
ax.plot(xvec,np.cos(xvec),'-b',label='cos(x)')

ax.legend()
ax.set_title('my funcions')
ax.set_xlabel('x')
ax.set_ylabel('cos(x)')
ax.grid(True)
```
