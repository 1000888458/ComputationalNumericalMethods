```python
import math as mt
def cos_tay(x,k):
  tay=0
  for i in range(0,k+1):
    tay=1-((x**2)/mt.factorial(2))+((x**4)/mt.factorial(4))
    result=tay
  return result
cos=cos_tay(0.2,2)
print("Resultado =",cos)
```

```python
import numpy as np
xvect=np.arange(-mt.pi,mt.pi,(mt.pi*2)/100)
cos_aprox=[]
for i in xvect:
  cos_aprox.append(cos_tay(i,3))
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,10))
ax.plot(xvect,cos_aprox,label="Serie de taylor")
ax.plot(xvect,np.cos(xvect),label="Cos(x)")
ax.legend()
```
