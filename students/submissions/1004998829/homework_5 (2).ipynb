```python
def b_coeffs(i,k,x,y):                          
    # i, k are the first and last INDEX of `x`
    #If i+k>N
    if i+k>=len(x):
        return 0
      #Zeroth divided difference
    elif k == 0:
        return y[i]
      #If higher divided difference
    else:
        return (b_coeffs(i+1, k-1, x, y)-b_coeffs(i, k-1, x, y))/(x[i+k]-x[i])
```

```python
def cuadratic_interp(x,y,xinterp):
    import numpy as np #asignamos los coeficientes al polinomio cuadratico
    b0 = b_coeffs(0,0,x,y)
    b1 = b_coeffs(0,1,x,y)  
    b2 = b_coeffs(0,2,x,y)
    yinterp = b0 + b1*(xinterp-x[0]) + b2*(xinterp-x[0])*(xinterp-x[1])
    return yinterp
```

```python
import numpy as np # Creamos los conjuntos de datos para testear la funcion de interpolacion
x=np.array([4,5,6])
y=np.array([np.log10(4),np.log10(5),np.log10(6)]) #vamos a interpolar la funcion de logaritmo base 10
val=np.arange(-1,21,0.1) #esto estaba originalmente como un comentario pero por curiosidad lo voy a usar luego para graficar


result=cuadratic_interp(x,y,val)#esto es lo que se usara luego para graficar
```

```python
cuadratic_interp(x,y,5.0) #aqui estamos interpolando la funcion de log10 en el valor "5"
```

```python
np.log10(5.0)==cuadratic_interp(x,y,5.0) #verificamos que si sea correcto
```
