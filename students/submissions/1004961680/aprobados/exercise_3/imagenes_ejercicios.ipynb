Ejercicio 3, Matriz de 25 imágenes (5x5) aleatorias de un conjunto de datos y volver a imprimir las imágenes en escala de grises

```python
import tensorflow as tf
(X_data, y_data), (_, _) = tf.keras.datasets.cifar10.load_data()
# X_data (numpy array) contiene los imagenes 
# y_data (numpy array) contiene las etiquietas
```

```python
X_data.shape 
```

```python
fig,ax = plt.subplots(5,5,figsize=(15,15)) 

for i in range(5):
    for j in range(5):
        
        random = int(np.random.randint(0,1797,1)) 
        
        ax[i,j].imshow(X_data[random]) 
        ax[i,j].set_title(y_data[random]) 
```

```python
fig,ax = plt.subplots(5,5,figsize=(15,15)) 

for i in range(5):
    for j in range(5):
        
        random = int(np.random.randint(0,1797,1))
        
        ax[i,j].imshow(np.average(X_data[random],-1)) 
        ax[i,j].set_title(y_data[random]) 
```
