**TAREA PARA NAVIDAD**

**f(x)=(x^2)-11**

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f_exp(x):
    return (x**2)-11

x=np.linspace(-5,5,100)
fig,ax = plt.subplots(1,1)
ax.plot(x,f_exp(x))
ax.grid('True')




fx = lambda x: (x**2)-11

a = 3
b = 4
tolera = 0.001


tramo = b-a
while not(tramo<tolera):
    c = (a+b)/2
    fa = fx(a)
    fb = fx(b)
    fc = fx(c)
    cambio = np.sign(fa)*np.sign(fc)
    if cambio < 0: 
        a = a
        b = c
    if cambio > 0:
        a = c
        b = b
    tramo = b-a


print('Tiene raiz en: ', c)
```

**Bisection method**

**f(x)=** = (e^-x)-x

```python




import numpy as np
import matplotlib.pyplot as plt
import math

def f_exp(x):
    return ((math.e)**-x)-x

x=np.linspace(-5,5,100)
fig,ax = plt.subplots(1,1)
ax.plot(x,f_exp(x))
ax.grid('True')




fx = lambda x: ((math.e)**-x)-x

a = 0
b = 2
tolera = 0.001
bis=[]


tramo = b-a
while not(tramo<tolera):
    c = (a+b)/2
    fa = fx(a)
    fb = fx(b)
    fc = fx(c)
    cambio = np.sign(fa)*np.sign(fc)
    
    if cambio < 0: 
        a = a
        b = c
    if cambio > 0:
        a = c
        b = b
    tramo = b-a
    
    bis.append([c,tramo])




print([ 'c', 'tramo'])
np.set_printoptions(precision = 4)
print(bis)
print('Tiene raiz en: ', c)


```
