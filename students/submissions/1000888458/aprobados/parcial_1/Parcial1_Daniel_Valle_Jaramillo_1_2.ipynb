```python
import numpy as np
import math

#1)
def logaritmo(x):
  suma=0
  contador=0
  
  for i in range(1,5):
    if x<=-1 and x>1:
      print("Error, rango equivocado")

    
    suma_log=((-1)**(i+1))*((x**i)/i)
    suma=suma_log+suma
    contador=contador+1

  return suma

print(logaritmo(0.5))

print(np.log(1+0.5))

```

```python
#2)
def log_taylor(x,k):
  sumatoria=0
  cont=0
  for i  in range(1,k):
    In=((-1)**(i+1))*((x**i)/i)
    sumatoria=In+sumatoria
    cont=cont+1
  return sumatoria

log_taylor(0.5,5)
```

```python
#3)
import numpy as np

xvec=np.linspace(-1,1,100)[1:]
log_approx=[]

for x in xvec:
  resultado=log_taylor(x,5)
  log_approx.append(resultado)

print(log_approx)
```

```python
#4)
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
ax.plot(xvec,log_approx,'--r',label='Taylor Approximation')  
ax.plot(xvec,np.log(xvec+1),'--b',label='numpy')
ax.legend()
ax.grid('True')
```
