## 3. 
Cree una función llamada `root_finding()` que devuelva el valor de la raíz (si existiera) para cada parametro `c` (pense `Newton` o `bisecion`). Usando esta funcion vamos a buscar primero raiz izquierda y despues raiz derecha.

```python
# tu codigo:

import numpy as np
import matplotlib.pyplot as plt
import math
def root_finding(a,b,parametro):
  def f_cost(x,c):
    fx=((x+1)**2)-c*(x-1)

    return fx




  fx = lambda x: ((x+1)**2)-parametro*(x-1)

  
  tolera = 0.001


  tramo = b-a
  while not(tramo<tolera):
      d = (a+b)/2
      fa = fx(a)
      fb = fx(b)
      fc = fx(d)
      cambio = np.sign(fa)*np.sign(fc)
      if cambio < 0: 
        a = a
        b = d
      if cambio > 0:
        a = d
        b = b
      tramo = b-a


  return d


root_finding(1,2,10),root_finding(6,7,10),root_finding(1,2,15),root_finding(11,12,15),root_finding(1,2,20),root_finding(16,17,20)



  
```
