```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets


digits = datasets.load_digits() # digits is a dictionary

digits["data"]
np.shape(digits["data"])
         


plt.figure(dpi=1200)
fig, ax = plt.subplots(5,5,figsize=(10,10))
for i in range(5):
    for j in range(5):
        ran = np.random.choice(digits["target"]) 
        
        numbers = np.reshape(digits["data"][ran],(8,8))
        ax[i,j].imshow(numbers)
        ax[i,j].set_title(ran)
```
