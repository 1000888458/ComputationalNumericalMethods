```python
#Ejercicio 1: Serie de Taylor

import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#Punto 1

x=0.2
k=2
cos=0

for n in range(k):
    cos += ((-1)**n)*(x**(2*n))/(math.factorial(2 * n))


print(cos)

```

```python
#Punto 2

def cos_taylor(x, k):
    cos = 0
    for n in range(k):
        cos += ((-1)**n)*(x**(2*n))/(math.factorial(2 * n))
    return cos
```

```python
#Punto 3

cos_aprox=[]
xvec = np.linspace(-np.pi,np.pi,100)

for i in np.arange(100):
  q = cos_taylor(xvec[i], 3)
  cos_aprox.append(q)

print(cos_aprox) 
```

```python
#Punto 4

plt.figure(figsize=(10,6))
plt.plot(xvec,cos_aprox,marker='*',color='red',linestyle='--',label="Cos Aprox.")
plt.plot(xvec,np.cos(xvec),marker='+',color='blue',linestyle='-.', label="Np Cos")

plt.grid()
plt.legend()
plt.show()

```
