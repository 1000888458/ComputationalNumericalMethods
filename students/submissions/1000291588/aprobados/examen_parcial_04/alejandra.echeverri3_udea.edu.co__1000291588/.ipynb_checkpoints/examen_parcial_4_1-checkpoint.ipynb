## Ejercicio 3.  $\color{red}{\text{(3p)}}$ Equaciones diferenciales - problema de valor de limite (boundary value problem).

Los organizadores de un espectáculo de fuegos artificiales quieren que los fuegos exploten a 40 metros sobre el suelo 5 segundos después del lanzamiento. Encuentra la velocidad inicial del cohete. 

a) usar ecuacion diferencial segunda orden:

$$\frac{d^2y}{dt^2}= −g $$

donde $y$ es altitud de cohete y,   

b) hacer plot $y(t)$ vs $t$


````python
#a)
from scipy.integrate import solve_bvp
import numpy as np

def derivs(t,y):
    dydt=np.zeros((y.shape))
    dydt[0]=y[1]
    dydt[1]=-9.8
    return dydt


def bc(ya,yb):
    residuals = np.array([0.0,0.0])
    residuals[0]=ya[0] - 0.0
    residuals[1]=yb[0] - 40.0 #condition
    return residuals

# give the time domain
t = np.linspace(0,5,100)

y0 = np.ones((2,100))


sol = solve_bvp(derivs, bc, t, y0)
sol.y[1][-1]

print(f'La velocidad incial del cohete es: V = {sol.y[1][0]} m/s')
````

Veamos como se comporta la velocidad con respecto a la posición

````python
plt.plot(sol.y[0], sol.y[1] )
plt.xlabel('Position')
plt.ylabel('Velocity')
plt.grid()
plt.title('v vs y')
````

````python
#b)
import matplotlib.pyplot as plt
fig, ax=plt.subplots(1,1)
ax.plot(sol.x, sol.y[0],color='darkcyan')
ax.set_xlabel('time')
ax.set_ylabel('y(t)')
ax.grid()
ax.set_title('y vs t')
````

## Ejercicio 4 Ecuaciones diferenciales acopladas - problema de valor inicial `scipy.integrate.solve_ivp` $\color{red}{\text{(10p)}}$ o `otros metodos`  $\color{red}{\text{(7p)}}$

Las ecuaciones de Lotka-Volterra, también conocidas como ecuaciones depredador-presa, son un par de ecuaciones de primer orden, no lineales, ecuaciones diferenciales utilizadas con frecuencia para describir la dinámica de los sistemas biológicos en los que interactúan dos especies, uno como depredador y el otro como presa. Las poblaciones cambian a través del tiempo de acuerdo con el par de ecuaciones:

$$\frac{dx}{dt}=x(\color{red}{\alpha}-\color{red}{\beta} y)$$
$$\frac{dy}{dt}=-y(\color{red}{\gamma}-\color{red}{\delta} x)$$

Esa ecuación describiría la evolución de la población de conejos y zorros (condiciones perfectas)

* x - numbero de conejos  
* y - numbero de zorros

* $\color{red}{\alpha, \beta, \gamma, \delta }$ son parámetros que describen la interacción de las dos especies

* $\frac{dx}{dt}, \frac{dy}{dt}$ representan las tasas de crecimiento de las dos poblaciones a lo largo del tiempo

* $t$ es tiempo

<img src="rabbits_foxes.png" width="300" />

a) solucionar los ecuaciones diferenciales usando `scipy.integrate.solve_ivp` $\color{red}{\text{(10p)}}$ o `otros metodos`  $\color{red}{\text{(7p)}}$
asumir: 
* $\alpha, \beta, \gamma, \delta = 1$
* $x_{ini}=4$
* $y_{ini}=2$
* $t=0:30$ (unidades arbitrarias, asumir meses)

b) hacer plot con dos curvas: $x$ vs. $t$ , $y$ vs. $t$   
c) hacer plot $x vs y$  
d) comentar sobre los dos plots  

````python
from scipy import integrate
#parametros y condiciones iniciales
alpha = 1 
beta = 1
delta = 1
gamma = 1
x0 = 4
y0 = 2

#a)
def derivative(X, t, alpha, beta, delta, gamma):
    x, y = X
    equiation1_x = x * (alpha - beta * y)
    equiation2_y = -y * (gamma - delta*x)
    V = [equiation1_x, equiation2_y]
    return V

t = np.linspace(0,30,1000)
X0 = [x0, y0] #vector con condiciones iniciales
res = integrate.odeint(derivative, X0, t, args = (alpha, beta, delta, gamma))

# b) plot
x, y = res.T
plt.figure(figsize=(10,4))
plt.title("X vs t and Y vs t")
plt.plot(t, x, label = 'Conejos', color='darkcyan')
plt.plot(t, y, label = 'Zorros', color='red')
plt.xlabel('Time t (meses)')
plt.ylabel('Population x and y')
plt.legend()
plt.grid()

plt.show()
````

````python
#c)

plt.plot(x,y, color='gray')
plt.xlabel('conejos')
plt.ylabel('zorros')
plt.grid()
````

De la primera grafica claramente se puede observar cómo se correlacionan las poblaciones de ambas especies, pues claramente se ve que mientras la poblacion de zorros crece la poblacion de conejos disminuye. Tambien se puede ver que si la poblacion de zorros es muy baja, entonces la poblacion de conejos es muy alta, esto porque hay muy pocos depredadores. Tambien podemos observar una periodicidad en mabas poblaciones. Tambien podemos decir que el comportamiento de ambas especies es el mismo, pero que su poblacion sea mazima o minima depende de la otra especie, no se comportan a la par.

En la segunda grafica se conftima lo que se mencionó con anterioridad, pues esa curva cerrada ilustra que el sistema es periodico y que ambas poblaciones estan estrechamente relacionadas.
