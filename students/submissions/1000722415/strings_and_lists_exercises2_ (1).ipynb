### Siguiendo el material de la clase pasada hacer un plot de occurencia de un caracter (Jim o Tom o Huck) de libro atraves de los capitulos


```python
chapters_of_the_book=book.split('CHAPTER')
print(chapters_of_the_book[37])
print(len(chapters_of_the_book))
```

```python
import numpy as np

tom=[]
jim=[]
huck=[]

for i in range(36, len(chapters_of_the_book)):
  tom.append(chapters_of_the_book[i].count('Tom') + chapters_of_the_book[i].count('TOM'))
  jim.append(chapters_of_the_book[i].count('Jim') + chapters_of_the_book[i].count('JIM'))
  huck.append(chapters_of_the_book[i].count('Huck') + chapters_of_the_book[i].count('HUCK'))
 
 
chapters=np.linspace(1, 35, 35)
```

```python
import matplotlib.pyplot as plt
```

```python
fig,ax=plt.subplots(1,1,figsize=(10,5))
ax.plot(chapters,tom,'-r',label='tom')
ax.plot(chapters,jim,'-b',label='jim')
ax.plot(chapters,huck,'-g',label='huck')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.05)
ax.set_xlabel('Chapters')
ax.set_ylabel('Characters Occurrences ')
ax.set_title('Characters Occurrences in each chapter')
ax.grid()

print('The character with most ocurrences is Tom with ',np.sum(tom), 'ocurrences')
print('The character with less ocurrences is Jim with ',np.sum(jim), 'ocurrences')
print(' ')
#fig.savefig('my_plot.png')
```
