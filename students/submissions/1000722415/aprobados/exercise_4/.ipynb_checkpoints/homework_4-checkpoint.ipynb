```python
import numpy as np
import matplotlib.pyplot as plt

def biseccion(f,a,b,tol):    
    p=(a+b)/2
    
    n=0

    while abs(f(p))>tol:

        if f(a)*f(p)<0:
            b=p
        if f(p)*f(b)<0:
            a=p

        p=(a+b)/2
        n+=1
    
    return p

def f(x):
  return x**2-11

a = 3
b = 8
tolera = 0.001

c=biseccion(f,a,b,tolera)

x=np.linspace(-10,10,100)
y=f(x)
plt.plot(x,y)
plt.scatter(c,0)
plt.title('Bisection Method')
plt.grid()
```

```python
import sympy 
def newton_raphson(f,dx,p0,tol):
    import sympy     
    n=0
    p=p0
    derivada=dx(p)
    while abs(f(p))>tol:
        p = p - f(p)/derivada
        n+=1
    return p
```

```python
def dx(x):
  return 2*x

tolerancia=0.001
p0=16

root=newton_raphson(f,dx,p0,tolerancia)

x=np.linspace(-10,10,100)
y=f(x)
plt.plot(x,y)
plt.scatter(root,0)
plt.title('Newton-Raphson Method')
plt.grid()
```
