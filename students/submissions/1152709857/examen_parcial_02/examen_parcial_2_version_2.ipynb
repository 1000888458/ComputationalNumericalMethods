Para un dispositivo óptico particular que consiste en un medio no lineal dentro de una cavidad óptica, esto sólo ocurre cuando la intensidad I se encuentra dentro de un cierto rango que está dado por la ecuación:

## 1. 
Cree una función llamada `f_cost` que reciba dos parámetros (x,c) y solo devuelva la función f(x) como se mencionó anteriormente.

```python
# tu codigo:
def f_cost(x,c):
  return ((x+1)**2) -c*(x-1)
```

## 2a. Para cual `c` esta funcion no tiene raices?

```python
# tu respuesta 
#Defining Bisection function
def Bisection( f, a, b, c, Nmax, printer=False ):
    #verifying the STEP1, a and b with different signs
    if f(a,c)*f(b,c)>0:
        print ("Error, f(a) and f(b) should have opposite signs")
        return False
    #Assigning the current extreme values, STEP2
    ai = a
    bi = b
    #Iterations
    n = 1
    while n<=Nmax:
        #Bisection, STEP3
        pi = (ai+bi)/2.0
        #Evaluating function in pi, STEP4 and STEP5
        if printer:
            print ("Value for %d iterations:"%(n),pi)
        #Condition A
        if f(pi,c)*f(ai,c)>0:
            ai = pi
        #Condition B
        elif f(pi,c)*f(ai,c)<0:
            bi = pi
        #Condition C: repeat the cycle
        n+=1
    #Final result
    return pi

#aqui variamos los rangos a y b para buscar las raices
a=0 
b=5
Nmax=10
print("Para c=5")
result1=Bisection(f_cost,a,b, c1, Nmax, True)
print("para c=10")
result2=Bisection(f_cost,a,b, c2, Nmax, True)
print("Para c=15")
result3=Bisection(f_cost,a,b, c3, Nmax, True)
print("Para c=20")
result4=Bisection(f_cost,a,b, c4, Nmax, True)

#observemos que la funcion para c=5 no tiene raices
```

## 3. 
Cree una función llamada `root_finding()` que devuelva el valor de la raíz (si existiera) para cada parametro `c` (pense `Newton` o `bisecion`). Usando esta funcion vamos a buscar primero raiz izquierda y despues raiz derecha.

```python
# tu codigo:
def root_finding( f, a, b, c, Nmax, printer=False ):
    #verifying the STEP1, a and b with different signs
    if f(a,c)*f(b,c)>0:
        print ("Error, f(a) and f(b) should have opposite signs")
        return False
    #Assigning the current extreme values, STEP2
    ai = a
    bi = b
    #Iterations
    n = 1
    while n<=Nmax:
        #Bisection, STEP3
        pi = (ai+bi)/2.0
        #Evaluating function in pi, STEP4 and STEP5
        if printer:
            print ("Value for %d iterations:"%(n),pi)
        #Condition A
        if f(pi,c)*f(ai,c)>0:
            ai = pi
        #Condition B
        elif f(pi,c)*f(ai,c)<0:
            bi = pi
        #Condition C: repeat the cycle
        n+=1
    #Final result
    return pi

#raiz izquierda
print("raiz izquierda")
a=0 
b=5
Nmax=10
print("Para c=5")
result1=Bisection(f_cost,a,b, c1, Nmax, True)
print("para c=10")
result2=Bisection(f_cost,a,b, c2, Nmax, True)
print("Para c=15")
result3=Bisection(f_cost,a,b, c3, Nmax, True)
print("Para c=20")
result4=Bisection(f_cost,a,b, c4, Nmax, True)

print()
#raiz derecha
print("raiz derecha")
a=5 
b=20
Nmax=10
print("Para c=5")
result1=Bisection(f_cost,a,b, c1, Nmax, True)
print("para c=10")
result2=Bisection(f_cost,a,b, c2, Nmax, True)
print("Para c=15")
result3=Bisection(f_cost,a,b, c3, Nmax, True)
print("Para c=20")
result4=Bisection(f_cost,a,b, c4, Nmax, True)
```

```python
# tu codigo:
c_vec=np.linspace(5,100,50)
left_roots=[]
for i in c_vec:
  a=0 
  b=5 
  Nmax=10
  result=root_finding(f_cost,a,b,i,Nmax)
  left_roots.append(result)
print(left_roots)
```
