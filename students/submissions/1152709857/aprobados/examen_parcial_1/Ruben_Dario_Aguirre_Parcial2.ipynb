```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#1. usando loop for calcular valor de Ln(1+x) con x=0.5 para k=1,2,3,4,5

x = 0.5
log = 0
for k in range(1,5):
  log += (-1)**(k+1)*(x**k)/k   
print("Ln(1+x) con la aproximacion por sumas =",log)
print("Ln(1+x) con numpy =",np.log(1+x))
```

```python
#2. crear una funcion log_taylor(x,k) que acepta dos parametros x y k y devuelve valor de Ln(1+x) para dado numero de terminos k

def log_taylor(x,k):
  log = 0
  for k in range(1,5):
    log += (-1)**(k+1)*(x**k)/k
  return log

log_taylor(0.5,5)  
```

```python
#3. crear un vector xvec que tiene 100 elementos en rango de -1 a 1sando loop for para cada elemento de este vector llama la funcion log_taylor(x,k) y guarda resultados en una lista log_approx

xvec = np.linspace(-1,1,101)[1:] #ponemos 101 dado que el primer termino no se toma
log_approx = []
for i in xvec:
  logn = log_taylor(i,5)
  log_approx.append(logn)

print(xvec)
print(log_approx)
```

```python
#hacer visualizacion de dos curvas en mismo plot
plt.figure(figsize=(14,7))
plt.title("Grafica ln(1+x)")
plt.plot(xvec, np.log(1+xvec), label='numpy')
plt.plot(xvec,log_approx,label='Aproximacion por sumas')
plt.xlabel("xvec")
plt.grid()
plt.ylabel("ln(1+x)")
plt.legend()
plt.show()
```
