```python
import matplotlib.pyplot as plt
import numpy as np
import math as math
```

### 1

```python
sum = 0
x = 0.2
for k in range(3):
    sum = (((-1)**k)*(x**(2*k)))/math.factorial(2*k) + sum
    #print(sum)
    
print(sum)
```

### 2

```python
def cos_taylor(x, k):
    sum = 0
    for i in range(k+1):
        sum = (((-1)**i)*(x**(2*i)))/math.factorial(2*i) + sum
        #print(i, sum)
        
    return(sum)
```

```python
cos_taylor(0.2,2)
```

### 3

```python
xvec = list(np.linspace(-np.pi, np.pi,100))
cos_approx=[]
for x in xvec:
    cos_approx.append(cos_taylor(x,3))
print(xvec)
```

### 4

```python
fig, ax = plt.subplots()
l1, = ax.plot(xvec,cos_approx)
l2, = ax.plot(xvec,np.cos(xvec))

ax.legend((l1, l2), ('Taylor approximation', 'cos(x)'), shadow=True)
plt.grid(True)
ax.set_xlabel('xvec')
ax.set_ylabel('Value')
ax.set_title('Taylor approximation comparison')
plt.show()
```
