````python
S = np.matrix([[1, -1, -1, -1],[-1, 2, 0, 0],[-1, 3, 3, 0],[-1, 0, 1, 4]])   
S
````

````python
def square(A):
    return A.shape[0] == A.shape[1]
````

````python
def matrix_exp(A):
    if square (A):
        eVals,eVecs = eig(A)    
        D = np.zeros((len(A),len(A)))
        E = np.zeros((len(A),len(A)))
        for i in range(0,len(eVals)):
            D[i,i] = eVals[i].real
            E[i,i] = np.e**D[i,i]
        Sinv = inv(eVecs)  
        
        return (eVecs, E, Sinv)
    else:
        return(print("Matrix must be square","",""))
````
