```python
#Libraries
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
def integrand(x,k):
    return 1/(1 -( k**2 * np.sin(x)**2))
def Period(K, l, g):
    return 4*np.sqrt(l/g)*K
   

l = 0.2 
g = 9.81
θ = [0,10,20,30,40,50,60]
k = np.sin(θ)/2

Ks = []
for i in θ:
    I, err = integrate.quad(integrand, 0, 2*np.pi, args =(i))
    Ks.append(I)
    
vPeriod = np.vectorize(Period)
T = vPeriod(Ks,l, g)

#Plot
plt.figure(figsize=(10, 6), dpi=80)
plt.plot(θ, T, color = "Red", linestyle = ":")
plt.ylabel("Period $(T)$")
plt.xlabel("Angle $(θ)$")
plt.title(" T vs θ")
plt.grid()
```
