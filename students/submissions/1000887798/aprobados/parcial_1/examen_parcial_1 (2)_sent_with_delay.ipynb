```python
import scipy.io
imgs_dict = scipy.io.loadmat('images_32x32.mat')
```

**2**

```python
print('Las llaves que contiene el diccionario son: ',imgs_dict.keys())
print('Las dimensiones de las llaves X y Y respectivamente son: ',imgs_dict['X'].shape,' ,  ', imgs_dict['y'].shape)
```

**3**

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

fig,ax = plt.subplots(1,10 ,figsize = (10,1))
for i in range(10):
  fila = np.random.randint(0,26032)
  img = Image.fromarray(imgs_dict['X'][:,:,:,fila].reshape(32,32,3))
  number = imgs_dict['y'][fila]
  ax[i].imshow(img, aspect = 'auto')
  ax[i].set_title(number)   
```

**4**

```python
prom_pix = (imgs_dict['X'][:,:,2,:] + imgs_dict['X'][:,:,1,:] + imgs_dict['X'][:,:,0,:])/3
prom_pix[4]
```

```python
fig,ax = plt.subplots(1,10 ,figsize = (10,1))
for i in range(10):
  fila = np.random.randint(0,26032)
  img = Image.fromarray(prom_pix[:,:,fila])
  number = imgs_dict['y'][fila]
  ax[i].imshow(img, aspect = 'auto')
  ax[i].set_title(number)   
```
