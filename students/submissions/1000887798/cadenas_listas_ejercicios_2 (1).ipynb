### Siguiendo el material de la clase pasada hacer un plot de occurencia de un caracter (Jim o Tom o Huck) de libro atraves de los capitulos


```python
chapters_of_the_book=book.split('CHAPTER')
print(chapters_of_the_book[37])
print(len(chapters_of_the_book))
```

```python
import numpy as np

tom=[]
jim=[]
huck=[]

for i in range(36, len(chapters_of_the_book)):
  tom.append(chapters_of_the_book[i].count('Tom'))
  jim.append(chapters_of_the_book[i].count('Jim'))
  huck.append(chapters_of_the_book[i].count('Huck'))

print(tom)
print(len(tom))
print(jim)
print(huck)   
```

```python
chapters=np.linspace(1,35,35)
print(chapters)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,1,figsize=(7,5))
ax.plot(chapters,tom,'-b',label='tom')
ax.plot(chapters,jim,'-g',label='jim')
ax.plot(chapters,huck,'-r',label='huck')
ax.legend(bbox_to_anchor=(1.05, 1), loc='center left', borderaxespad=0.05)
ax.set_xlabel('Numero de capitulos')
ax.set_ylabel(' Repeticiones')
ax.set_title('Repeticiones de cada nombre por capitulo')
ax.grid()

print('El nombre que más se repite es Tom con ',np.sum(tom))
print('El nombre que menos aparece es Jim con ',np.sum(jim))
print(' ')

```
