```python
from sklearn import datasets
digits = datasets.load_digits() # digits is a dictionary

fig,ax = plt.subplots(5,5 ,figsize = (12,12))
for i in range(5):
  for j in range(5):
    fila = np.random.randint(0,1797)
    img = digits['data'][fila].reshape(8,8,)
    number = digits['target'][fila]
    ax[i,j].imshow(img, aspect = 'auto')
    ax[i,j].set_title(number)    
plt.savefig("8x8.jpg", bbox_inches='tight')



plt.gray()
```

```python
import tensorflow as tf
(X_data, y_data), (_, _) = tf.keras.datasets.cifar10.load_data()
# X_data (numpy array) contiene los imagenes 
# y_data (numpy array) contiene las etiquietas
```

```python
from PIL import Image
fig,ax = plt.subplots(5,5 ,figsize = (16,16))
for i in range(5):
  for j in range(5):
    fila = np.random.randint(0,50000)
    img = Image.fromarray(X_data[fila,:,:,:])
    name = y_data[fila,0]
    ax[i,j].imshow(img, aspect = 'auto')
    ax[i,j].set_title(name) 
```

```python
prom_pix = (X_data[:,:,:,2] + X_data[:,:,:,1] + X_data[:,:,:,0])/3
fig,ax = plt.subplots(5,5 ,figsize = (16,16))
for i in range(5):
  for j in range(5):
    fila = np.random.randint(0,50000)
    img = Image.fromarray(prom_pix[fila,:,:])
    name = y_data[fila,0]
    ax[i,j].imshow(img, aspect = 'auto')
    ax[i,j].set_title(name) 
```
