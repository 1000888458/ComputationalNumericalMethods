# **EJERCICIO 2**

````python
#punto 2
#graficamos temperature vs pressure
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data_pressure_vs_temparature (1).csv")
df.head()

import numpy as np
import matplotlib.pyplot as plt
temp= df['Temperature'].to_numpy()
pre= df['Pressure'].to_numpy()

fig, ax= plt.subplots(1,1,figsize=(7,5))
ax.plot(temp,pre,'--r')
ax.set_title('Temperature vs Pressure')
ax.set_xlabel('Temperature')
ax.set_ylabel('Pressure')
ax.grid(True)
````

````python
#punto c. derivada numerica y grafica de capacidad calorifica vs T
from scipy import gradient 
cv=np.gradient(pre, temp)

fig, ax= plt.subplots(1,1,figsize=(7,5))
ax.plot(temp,cv,'--r')
ax.set_title('Temperature vs Capacity')
ax.set_xlabel('Temperature')
ax.set_ylabel('Capacity')
ax.grid(True)
````

````python
#ejercicio d
from numpy import diff
x = df['Temperature']
y = df['Pressure']
dydx = diff(y)/diff(x)
dydx = np.insert(dydx,0,0)
print("la temperatura critica del gas es donde hay un pico o un cambio abrupto, por lo tanto su valor es:", max(dydx))
````
