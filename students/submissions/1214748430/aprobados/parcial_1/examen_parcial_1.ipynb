```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

### Ejercicio 1a)

__Primer punto:__

```python
sum = 0
for i in range(3):
    sum += (-1)**i*(0.2)**(2*i)/math.factorial(2*i)
print(sum)
```

__Segundo punto:__

```python
def cos_taylor(x,k):
    sum = 0
    for i in range(k+1):
        sum += (-1)**i*(x)**(2*i)/math.factorial(2*i)
    return(sum)
```

```python
cos_taylor(np.pi,7)
```

__Tercer punto:__

```python
xvec = np.linspace(-np.pi, np.pi, 100)
cos_approx=[]
for x in xvec:
    cos_approx.append(cos_taylor(x,3))
print(cos_approx)
    
```

__Cuarto punto:__

```python
fig,ax = plt.subplots(1,1)
#fig,ax=plt.subplots(1,1,figsize=(6,4))
ax.plot(xvec,cos_approx,label='3th approx')
ax.plot(xvec,np.cos(xvec),label='cos(x)')
ax.plot(x,x-x**3,label='1th approx')
ax.set_ylim(-1.1,1.1)
plt.legend()
plt.grid()
```
