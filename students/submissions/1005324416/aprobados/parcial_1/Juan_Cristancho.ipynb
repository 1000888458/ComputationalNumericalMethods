```python
#2
import math

def cos_taylor(x,k):
    x=x
    suma=0
    for k in range(0,k+1):
        sumatoria= ((-1)**k)*(x**(2*k))/math.factorial(2*k)
        suma=suma+sumatoria
    print(suma)
    return x,k

x,k = cos_taylor(0.2,2)
```

```python
#3
import numpy as np
import math

#x,k = cos_taylor(0.2,2)  #llama la funcion del ejercicio anterior

xvec=np.linspace(-np.pi,np.pi,100)
cos_appro=[]
for i in xvec:
    x=i
    suma=0
    for k in range(0,4):
        sumatoria= ((-1)**k)*(x**(2*k))/math.factorial(2*k)
        suma=suma+sumatoria
    cos_appro.append(suma)
    
print(cos_appro)
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi,np.pi,100)
y1 = cos_appro
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x,y1,'--y',label='Taylor Approximation')
ax.plot(x,y2,label='cos(x)')
ax.legend()
ax.set_title('my funcions')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
fig.savefig('my_plot.png')
```
