```python
#Metodo de biseccion 

import numpy as np
import matplotlib.pyplot as plt

#Funcion:
def f_cost(x):
    return x**2-11

#Datos:
xl=3 #x izquierda
xr=4 #x derecha
tol = 0.01
sol=[]

#Proceso:
diferencia= xr-xl
while not (diferencia < tol):
  xn = (xr+xl)/2
  signo = f_cost(xl)*f_cost(xn)
  if signo<0:
    xr=xn
  if signo>0:
    xl=xn
  diferencia = xr - xl
  sol.append(xn)

#Resultado
print("root: ", xn)
print("Lista de la evolucion: ", sol)


fig,ax = plt.subplots(1,1)
ax.plot(sol)
ax.grid('True')
plt.xlabel("intento")
plt.ylabel("Root")
```

```python
#metodo de newton rapson
import numpy as np
import matplotlib.pyplot as plt

#Funcion y derivada
def f_cost(x):
    return x**2-11
def df_cost(x):
    return 2*x

xr = 4
tol = 0.01

#Proceso
sol = []
diferencia = abs(2*tol)
while (diferencia>=tol):
    xn = xr - f_cost(xr)/df_cost(xr)
    diferencia  = abs(xn-xr)
    sol.append(xn)
    xr = xn

print("raiz en: ", xr)
print("Evolucion de la aproximacion: ", sol )

fig,ax = plt.subplots(1,1)
ax.plot(sol)
ax.grid('True')
plt.xlabel("intento")
plt.ylabel("Root")
```
