```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

## Primer punto

```python
#Graficación para diferentes metalicidades
    
#Cargando datos    
m00=np.loadtxt("m0.0")
m_05=np.loadtxt("m-0.5")
m05=np.loadtxt("m0.5")
m10=np.loadtxt("m-10")
m15=np.loadtxt("m-15")
m20=np.loadtxt("m-20")
m30=np.loadtxt("m-30")
mnil=np.loadtxt("mnil")

#T[Temperatura,metalicidad] Coo=[Función de enfriamiento,metalicidad] están en orden descendente
T=np.array([(m05[:,0],0.5),(m00[:,0],0.0),(m_05[:,0],-0.5),(m10[:,0],-10),(m15[:,0],-15),(m20[:,0],-20),(m30[:,0],-30),(mnil[:,0],"nil")])
Coo=np.array([(m05[:,5],0.5),(m00[:,5],0.0),(m_05[:,5],-0.5),(m10[:,5],-10),(m15[:,5],-15),(m20[:,5],-20),(m30[:,5],-30),(mnil[:,5],"nil")])

#Cantidad de vectores que tiene T, es decir, cantidad de metalicidades.
n=len(T[:,0])

plt.figure( figsize=(7,5) )

#Ploteo 
for i in xrange(n):
    plt.plot(T[i,0],Coo[i,0], linewidth=2, label="m="+str(T[i,1]))    
      
plt.grid()
plt.title("Cooling functions for different metallicities")
plt.xlabel("$log10(T) [k]$")
plt.ylabel("$log10[\Lambda(T)]\hspace{0.5} [ergs \hspace{0.3}cm^3\hspace{0.3} s^{-1}]$")
plt.legend()
plt.show()
```

# Segundo punto

```python
#Función de Interpolación lineal
def LinearInterpolator( x, Xn, Yn ):
    
    #Ordenamiento de datos, por si no lo están.
    Yn = Yn[np.argsort(Xn)]
    Xn = Xn[np.argsort(Xn)]
    
    #Hallando el tamaño de n. Si no es un arreglo se convierte en tal.
    try:
        Ninter = len(x)
    except:
        Ninter = 1
        x = np.array([x,])
        
    #Ecuación de la linea recta que pasa por dos puntos.
    term = lambda x, i: (Yn[i+1] - Yn[i])/(Xn[i+1] - Xn[i])*(x - Xn[i]) + Yn[i]
    
    #Detección de los intervalos para x [x_i,x_i+1], e interpolación.
    for n in xrange(Ninter):
        for i in xrange(len(Xn)):
            if x[n] <= Xn[i]:
                break
        y=term(x[n],i-1) 
    return y
```

```python
#Función CoolingFuction
def CoolingFunction(m,t):
    #Hallamos los valores de función de enfriamiento y Temperatura que correspondan a la metalicidad m. 
    Coo1=Coo[Coo[:,1]==m]
    T1=T[T[:,1]==m]
    #Llamamos la función Interpolación lineal
    return LinearInterpolator(t,T1[0,0],Coo1[0,0])
```

## Tercer punto

```python
#Defining Bisection function
def Bisection( f, a, b, Nmax=30, printer=False ):
    #verifying the STEP1, a and b with different signs
    if f(a)*f(b)>0:
        print "Error, f(a) and f(b) should have opposite signs"
        return False
    #Assigning the current extreme values, STEP2
    ai = a
    bi = b
    #Iterations
    n = 1
    while n<=Nmax:
        #Bisection, STEP3
        pi = (ai+bi)/2.0
        #Evaluating function in pi, STEP4 and STEP5
        if printer:
            print "Value for %d iterations:"%(n),pi
        #Condition A
        if f(pi)*f(ai)>0:
            ai = pi
        #Condition B
        elif f(pi)*f(ai)<0:
            bi = pi
        #Condition C: repeat the cycle
        n+=1
    #Final result
    return pi
```

```python
#Función Temperatura
def Temperatura(coo,to,m,a,b):
#Definimos la función 
    f= lambda t: CoolingFunction(m,t) - coo
    #Llamamos la función puntofijo
    return Bisection(f,a,b)
```

```python
#Hay que resaltar que los intervalos a,b deben ser muy cercanos al to ingresado, para que pueda funcionar Bisección.
Temperatura(-22.99,6.5,-10,6.4,6.6)
```

## Cuarto punto

```python
#Figura de funciones de enfriamiento para diferentes metalicidades, de acuerdo a la temperatura.
Tem=np.array([4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0])

plt.figure( figsize=(12,5) )

for k in Tem:
    Y=[]  #Cooling function
    M=[]  #Metalicity 
    for i in xrange(len(T)):
        for j in xrange(len(T[i,0])):
            if T[i,0][j]==k:
                if Coo[i,1]!="nil":
                    M.append(Coo[i,1])
                    Y.append(Coo[i,0][j])
    #Ploteo                
    plt.plot(M,Y, 'o-', label="t="+str(k)) 
    plt.legend()

plt.grid()    
plt.title("Cooling functions for different metalicities")
plt.xlabel("$Metalicitiy$")
plt.ylabel("$log10[\Lambda(T)]\hspace{0.5} [ergs \hspace{0.3}cm^3\hspace{0.3} s^{-1}]$")
plt.show()
```

```python
#Función CoolingFunctionMetallicity
def CoolingFunctionMetallicity(m,t):
    #Arreglo con los valores de metalicidad, (Xn)
    Met=T[:,1][:7]
    #Función de enfriamiento (Yn)
    Y=[CoolingFunction(i,t) for i in Met]
    Y=np.array(Y)
    #Interpolación
    return LinearInterpolator(m,Met,Y) 
```
