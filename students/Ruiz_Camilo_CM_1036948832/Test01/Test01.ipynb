# Cooling Function

Linear Interpolator: Se decidio usar el interpolador lineal ya que el tiempo de computo requerido para hacer la interpolacion es considerablemente menor en comparacion con el tiempo requerido con el interpolador de Lagrange.

```python
#Linear Interpolating Function
def LinearInterpolator( x, Xn, Yn ):
    
    #Sorting data, in case they are not
    Yn = Yn[np.argsort(Xn)]
    Xn = Xn[np.argsort(Xn)]
    
    #Detecting size of x
    try:
        Ninter = len(x)
    except:
        Ninter = 1
        x = np.array([x,])
        
    #Constructing function for each iteration
    term = lambda x, i: (Yn[i+1] - Yn[i])/(Xn[i+1] - Xn[i])*(x - Xn[i]) + Yn[i]
    
    #Detecting intervals for each x, [x_i, x_i+1] and interpolating
    y = []
    for n in xrange(Ninter):
        for i in xrange(len(Xn)):
            if x[n] <= Xn[i]:
                break
        y.append( term(x[n],i-1) )
    
    return np.array(y)
```

Esta funcion devuelve cualquier valor de Cooling Function dandole valores conocidos de temperatura y cooling function.

```python
def CoolingFunction(t,m):
    T = m[:,0]
    C = m[:,5]

    CF = LinearInterpolator(t,T,C)
    
    return CF
```

# Temperature

Find root-solver, a continuacion se define la funcion $\Lambda(T;m) - \Lambda_{0} = 0$ la cual se resuelve por el metodo de biseccion el cual es mas preciso que punto fijo con este conjunto de datos, es necesario aclarar que para el uso correcto de biseccion el intervalo en el que se desea hallar la raiz debe de ser muy acotado para que este funcione apropiadamente. 

```python
def CF(T,M,CFo):
    CF = CoolingFunction(T,M) - CFo
    
    return float(CF)
```

```python
#Defining Bisection function
def Bisection( f, a, b, Nmax, m, CF, printer=False ):
    #verifying the STEP1, a and b with different signs
    if f(a,m,CF)*f(b,m,CF)>0:
        print "Error, f(a) and f(b) should have opposite signs"
        return False
    #Assigning the current extreme values, STEP2
    ai = a
    bi = b
    #Iterations
    n = 1
    while n<=Nmax:
        #Bisection, STEP3
        pi = (ai+bi)/2.0
        #Evaluating function in pi, STEP4 and STEP5
        if printer:
            print "Value for %d iterations:"%(n),pi
        #Condition A
        if f(pi,m,CF)*f(ai,m,CF)>0:
            ai = pi
        #Condition B
        elif f(pi,m,CF)*f(ai,m,CF)<0:
            bi = pi
        #Condition C: repeat the cycle
        n+=1
    #Final result
    return pi
```

A continuacion se define la funcion $\textbf{Temperature}$ la cual nos provee cualquier temperatura dandole los datos ya previamente conocidos de metalicidad y cooling function.

```python
def Temperature(CFo,m,Ta,Tb):
    T = Bisection(CF,Ta,Tb,30,m,CFo)
    
    return T
```
