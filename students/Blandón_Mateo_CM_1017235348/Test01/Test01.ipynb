2 . (20%) Write a routine called CoolingFunction that, given any logarithmic temperature value log10(T) and a metallicity and using the provided data, interpolates (using either Linear or Lagrange interpolation) the value of the respective cooling function value.

```python
def CoolingFunction (x,m):
    #Especificando datos a usar
    global Xn,Yn
    if m==05:
        Xn=x1
        Yn=m051
    elif m==00:
        Xn=x2
        Yn=m00
    elif m==-05:
        Xn=x2
        Yn=m052
    elif m==-10:
        Xn=x2
        Yn=m10
    elif m==-15:
        Xn=x2
        Yn=m15
    elif m==-20:
        Xn=x2
        Yn=m20
    elif m==-30:
        Xn=x2
        Yn=m30
    elif m==-40: #Numero asignado para nil; metalicidad nula.
        Xn=x2
        Yn=mnil
    
    #Constructing function for each iteration
    term = lambda x, i: (Yn[i+1] - Yn[i])/(Xn[i+1] - Xn[i])*(x - Xn[i]) + Yn[i]
    

    #for n in xrange(Ninter):
    for i in xrange(len(Xn)):
        if x<=Xn[i]:
             break
    y=term(x,i-1)
    
    return y
    
    
print CoolingFunction(5,00)
```

```python
def Temperature(tguess,m,yi):
    #Bucamos los datos correspondientes a la Metalicidad dada.
    if m==05:
        Xn=x1
        Yn=m051
    elif m==00:
        Xn=x2
        Yn=m00
    elif m==-05:
        Xn=x2
        Yn=m052
    elif m==-10:
        Xn=x2
        Yn=m10
    elif m==-15:
        Xn=x2
        Yn=m15
    elif m==-20:
        Xn=x2
        Yn=m20
    elif m==-30:
        Xn=x2
        Yn=m30
    else:
        Xn=x2
        Yn=mnil
        

    y=CoolingFunction(tguess,m)-yi #Este valor se ingresa en la rutina de bisección
    return y
```

```python
def Biseccion (f,a,b,Nmax,m,yi): #para elegir a y b trate de que a sea menor que b y el intervalo sea reducido.
    if f(a,m,yi)*f(b,m,yi)>0: #se añaden los parapetros necesarios.
        print "Se requiere que los extremos sean de distinto signo"
        return False
    else:
        ai = a
        bi = b
        n=0
        bpi=[]
        for n in xrange(Nmax):
            pi=(ai+bi)/2.0
            bpi.append(pi) #Se añade a la lista el punto medio del intervalo.
            if f(pi,m,yi)*f(ai,m,yi)>0:
                ai=pi
            elif f(pi,m,yi)*f(bi,m,yi)>0:
                bi=pi
    return bpi
```


Finally, write a routine called CoolingFunctionMetallicity that, given any metallicity value and any temperature value, calculates the respective Cooling function.
