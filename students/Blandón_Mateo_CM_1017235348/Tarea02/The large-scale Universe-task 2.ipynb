1. The Press–Schechter formalism: the number of objects of a mass range within a simulation of a portion of the Universe can be approximately predicted by the Press–Schechter formalism. The number of object with a mass between M and M+dM is:

$N(M)dM=1π√(1+n3)ρ¯/M^2(M/M∗)^{(3+n)}/6exp[−(M/M∗)^{(3+n)/3}]VdM$

where ρ¯ is the mean matter density of the Universe, n is the index of the power spectrum, V the volume of the simulation and M∗ is a critical mass parameter.

This scrip do the next:

Load the catalog of dark matter halos.

Using Matplotlib, generate and plot a histogram of log10(mass) for the halos, # i.e., how many halos are per logarithmic mass range.

Plot in the same window the theoretical prediction of the Press-Schechter formalism using the next parameters:

$N(M)=1.018×1014 M−0.889×exp(−9.7×10−32 M0.222)$


```python
%matplotlib
import numpy as np  #importa las librerias mathematicas y de graficacion
import matplotlib.pyplot as plt 
```

```python
d=np.loadtxt("dathalos.dat")#carga los datos del catalogo
```

```python
x=d[:,0] #toma todos los elementos de la primera columna del catalogo,en este caso tambien se puede hablar de una matriz A a la que se le toman los A[1][0]. LA COMA ES COMPLETAMENTE NECESARIA PARA TOMAR TODOS LOS ELEMENTOS. SE REPITE PARA LAS OTRAS COORDENADAS Y LA MASA.  
y=d[:,1]
z=d[:,2]
m=d[:,3]
```

```python
#Sacamos logaritmo base 10 a las masa:
m=np.log10(m)
```

```python
#hacemos un histograma con los logaritmos de las masas y otro en la misma ventana de The Press-Schechter formalism:

plt.figure() #crea figura

plt.subplot( 1, 2, 1 ) #crea en la figura una fila y dos columnas para dos graficas

plt.hist(m, bins=200, color="blue") #primer histograma

plt.grid(True) #pone cuadricula

#X label
plt.xlabel( "$log10(M)$" )

#Y label
plt.ylabel( "Number of objects", fontsize=20 )

plt.subplot( 1, 2, 2 ) #segundo histgrama

N=1.018e+14*m**(-0.889)*np.exp(-9.7e-32*m**0.222)

plt.hist(N, bins=200, color="red" )

plt.grid(True) #pone cuadricula

plt.savefig("Dark-Matter-Halos-Histogram.png") #guarda la imagen en formato png

from IPython.core.display import Image 
Image(filename='/home/mateo/Descargas/Dark-Matter-Halos-Histogram.png')
```

2.Distribution of massive halos: the more massive halos are generally located within high density regions of the large-scale distribution, like clusters and filaments. This is a consequence of the hierarchical structure formation, where small structures formed first and larger structures are made of the smaller ones.

Create a Python script that does the next:

Load the catalog of dark matter halos.

Select the more massive halos, namely $M>1×10^13 M⊙$.

Plot in a 3D scatter both, the complete catalogue of halos and the more massive halos (with different colours).

```python
%pylab inline   
from mpl_toolkits.mplot3d import Axes3D   #importa la graficadora 3D
```

```python
k1=d[d[:,3]>1e+13] #saca lo elementos cuya columna 4 sea mayor qque 10**13
k2=d[d[:,3]<1e+13]  #saca lo elementos cuya columna 4 sea menor qque 10**13

fig=plt.figure( figsize = (11,11) ) #crea fig
ax = fig.add_subplot(111, projection='3d')

X1=k1[:,0]
Y1=k1[:,1]  #coordenadas de k1
Z1=k1[:,2]

X2=k2[:,0]
Y2=k2[:,1]  #coordenadas de k2
Z2=k2[:,2]

ax.set_xlim(0,250)
ax.set_ylim(0,250)  #limites de los ejejs
ax.set_zlim(0,250)
ax.grid(True)  #cuadricula
ax.set_xlabel("Mpc")   #etuiquetas
ax.set_ylabel("Mpc")
ax.set_zlabel("Mpc")

ax.scatter( X1, Y1, Z1, s=0.8, linewidth=0.01, color="red", alpha=1) #grafica, alpha es transparencia
ax.scatter( X2, Y2, Z2, s=0.1, linewidth=0.01, color="blue", alpha=0.5)
plt.savefig("3D matter.png")
```
