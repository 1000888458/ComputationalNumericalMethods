## 1 - Formalismo Press–Schechter

```python
import numpy as np
import matplotlib.pyplot as plt

Datos = np.loadtxt("FOF.dat")

#Masa de los halos
M=Datos[:,3];
Mlog = np.log10(M);

#Funcion de Press-Schechter
Nm=1.018E14*M**(-0.889)*np.exp(-9.7E-32*M**0.222);


# Gráfico del histograma
plt.figure( figsize = (10,10) )
Bin,F,patch = plt.hist(Mlog,200,alpha=0.2,fc='m')
plt.plot(Mlog,Nm,'b-',lw=2)
plt.xlabel("Masa de los halos(escala logaritmica)")
plt.ylabel("Numero de objetos")
plt.savefig("Histograma.png")
```

```python
from IPython.core.display import Image 
Image(filename='Histograma.png')
```

Se puede ver en el grafico que la función de Press–Schechter se ajusta a la distribución de frecuencias de la masa de los halos. También se observa en la distribución, que la resolución de la simulación, asociada al valor más pequeño de la masa de los halos, es del orden de $10^{11}$.

##  2. Distribución de los Halos masivos

```python
#!/usr/bin/env python

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

#Gráficos en 3D
fig = plt.figure( figsize = (10,10) )
ax = fig.add_subplot(111, projection='3d')

# Carga de datos
x,y,z,M = np.loadtxt("FOF.dat", unpack=True)

#Selección los halos mas masivos
Bool = M>1E13
X = x[Bool]
Y = y[Bool]
Z = z[Bool]

#Gráfico 
ax.scatter(X,Y,Z,c='r',s=4,linewidth=0.05)
ax.scatter(x,y,z,s=0.15,linewidth=0.01,c='k')
ax.set_xlim([0,250])
ax.set_ylim([0,250])
ax.set_zlim([0,250])
plt.savefig("HalosXYZ.png")
```

```python
from IPython.core.display import Image 
Image(filename='HalosXYZ.png')
```

Se observa de la distribución de los halos masivos, de materia oscura, que efectivamente están ubicados en las regiones más densas.
