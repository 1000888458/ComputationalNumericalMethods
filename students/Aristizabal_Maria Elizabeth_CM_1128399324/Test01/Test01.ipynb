### 2. Función enfriamiento

```python
#Linear Interpolating Function
def LinearInterpolator( x, Xn, Yn ):
    
    #Sorting data, in case they are not
    Yn = Yn[np.argsort(Xn)]
    Xn = Xn[np.argsort(Xn)]
    
    #Detecting size of x
    try:
        Ninter = len(x)
    except:
        Ninter = 1
        x = np.array([x,])
        
    #Constructing function for each iteration
    term = lambda x, i: (Yn[i+1] - Yn[i])/(Xn[i+1] - Xn[i])*(x - Xn[i]) + Yn[i]
    
    #Detecting intervals for each x, [x_i, x_i+1] and interpolating
    y = []
    for n in xrange(Ninter):
        for i in xrange(len(Xn)):
            if x[n] <= Xn[i]:
                break
        y.append( term(x[n],i-1) )
    
    return np.array(y)
```

```python
#Definicion de la funcion enfriamiento
def CoolingFunction(x,Metalicidad):
    LogT=Metalicidad[:,0]
    LogL=Metalicidad[:,5]
    FE = LinearInterpolator(x,LogT,LogL)
    return FE
```

### 3. Función Temperatura

```python
#Defining Bisection function
def Bisection( f, a, b, Nmax, printer=False ):
    #verifying the STEP1, a and b with different signs
    if f(a)*f(b)>0:
        print "Error, f(a) and f(b) should have opposite signs"
        return False
    #Assigning the current extreme values, STEP2
    ai = a
    bi = b
    #Iterations
    n = 1
    while n<=Nmax:
        #Bisection, STEP3
        pi = (ai+bi)/2.0
        #Evaluating function in pi, STEP4 and STEP5
        if printer:
            print "Value for %d iterations:"%(n),pi
        #Condition A
        if f(pi)*f(ai)>0:
            ai = pi
        #Condition B
        elif f(pi)*f(ai)<0:
            bi = pi
        #Condition C: repeat the cycle
        n+=1
    #Final result
    return pi
```

```python
#Definicion de la funcion temperatura
def Temperatura(FE,metalicidad,Ta,Tb):
    Nmax=20
    
    #Se define la funcion para calcular los ceros.
    def func(x):
        return (CoolingFunction(x,metalicidad)-FE)
    
    resultado = Bisection(func,Ta,Tb,20)
    return resultado
```

```python
# Ejemplo:
# Para un valor de la funcion de enfriamiento igual a -21.5, una metalicidad de -0.5 y un intervalo (4.6, 5.0).
CFi=-21.5
Ta=4.6
Tb=5.0
Temperatura(CFi,M05,Ta,Tb)
```

### 4. Funciones de la metalicidad

```python
#Definicion de CoolingFunction para diferentes temperaturas
def CFTemperature(T):
    FeT = []
        
    for i in range(len(T)):
        FeT.append(CoolingFunction(T[i], M_05))
        FeT.append(CoolingFunction(T[i], M00))
        FeT.append(CoolingFunction(T[i], M05))
        FeT.append(CoolingFunction(T[i], M10))
        FeT.append(CoolingFunction(T[i], M15))
        FeT.append(CoolingFunction(T[i], M20))
        FeT.append(CoolingFunction(T[i], M30))
    FeT = np.array(FeT)
    
    return FeT
```

```python
#Datos
Metalicidades = np.array([0.5,0.0,-0.5,-1.0,-1.5,-2.0,-3.0])
T = np.arange(4.5,8.5,0.5)
cf = CFTemperature(T)
```

```python
#Gráfica
plt.figure(figsize(10,10))
plt.plot(Metalicidades,cf[0:7],label='$log(T)=4.5$')
plt.plot(Metalicidades,cf[7:14],label='$log(T)=5.0$')
plt.plot(Metalicidades,cf[14:21],label='$log(T)=5.5$')
plt.plot(Metalicidades,cf[21:28],label='$log(T)=6.0$')
plt.plot(Metalicidades,cf[28:35],label='$log(T)=6.5$')
plt.plot(Metalicidades,cf[35:42],label='$log(T)=7.0$')
plt.plot(Metalicidades,cf[42:49],label='$log(T)=7.5$')
plt.plot(Metalicidades,cf[49:56],label='$log(T)=8.0$')

plt.legend()
plt.title('Funcion de enfriamieto para diferentes temperaturas', fontsize = 15)
plt.xlabel('$Metalicidad$', fontsize = 15)
plt.ylabel('$Log 10 [\Lambda(T)] [ergs\, cm^3 s^{-1}]$', fontsize = 15)
plt.grid()
```

```python
#Definición de la función CoolingFunctionMetallicity
def CoolingFunctionMetallicity(T, m):
    Metalicidades = np.array([0.5,0.0,-0.5,-1.0,-1.5,-2.0,-3.0])
    cft = CFTemperature([T])
    Lambda = LinearInterpolator( m, Metalicidades, cft )
    return Lambda
```
