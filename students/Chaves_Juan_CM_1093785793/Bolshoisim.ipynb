```python
#Preliminares
import numpy as np
%pylab inline
import matplotlib.pyplot as plt
```

```python
#Importando los datos de Bolshoi
datos=np.loadtxt("FOF.dat")
#Extrayendo las masas
x=datos[:,3] #Extrae cuarta columna, donde se hallan las masas de los cumulos.
N=1.018*1e+14*x**-0.889*np.exp(-9.7*1e-32*x**0.222)
N=np.log10(N)
x=np.log10(x)
#Histograma Simulado
plt.title("Dark Matter Histogram")
plt.subplot(1,2,1)
plt.xlabel("$\log_{10}(Mass)$")
plt.ylabel("Frequency")
plt.grid(True)
plt.hist(x, log=True, bins=200, histtype='stepfilled', color='b', label='Simulated')
plt.legend(loc="upper right")
#Histograma teorico
plt.subplot(1,2,2)
plt.xlabel("$\log_{10}(Mass)$")
plt.ylabel("Frequency")
plt.grid(True)
plt.hist(N, log=True, bins=200, histtype='stepfilled', color='r', alpha=0.2, label='Predicted')
#plt.xscale('log') 
plt.legend(loc="upper right")
```

```python
from mpl_toolkits.mplot3d import Axes3D

smallhalos=datos[datos[:,3]<1e+13] #Seleccionando las coordenadas de cumulos tal que M<10**13 masas solares
X=smallhalos[:,0]
Y=smallhalos[:,1]
Z=smallhalos[:,2]

bighalos=datos[datos[:,3]>1e+13] #Seleccionando las coordenadas de cumulos tal que M>10**13 masas solares
X2=bighalos[:,0]
Y2=bighalos[:,1]
Z2=bighalos[:,2]

fig = plt.figure( figsize = (14,14) )
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(0,250)
ax.set_ylim(0,250)
ax.set_zlim(0,250)
ax.grid(True)
ax.set_xlabel("Mpc")
ax.set_ylabel("Mpc")
ax.set_zlabel("Mpc")

ax.scatter( X, Y, Z , color="grey", s=0.1, alpha=0.4) #s=size (tamaño de los puntos)
ax.scatter( X2, Y2, Z2 , color="red", s=0.8, alpha=1)
```
