**2**.Write a routine called CoolingFunction that, given any logarithmic temperature value $log_{10}(T)$ and a metallicity and using the provided data, interpolates (using either Linear or Lagrange interpolation) the value of the respective cooling function value.

```python
def CoolingFunction (x,m):
    global Xn,Yn
    #Especificando datos a usar
    if m==+5:
        Xn=x1
        Yn=m051
    elif m==0:
        Xn=x2
        Yn=m00
    elif m==-5:
        Xn=x2
        Yn=m052
    elif m==-10:
        Xn=x2
        Yn=m10
    elif m==-15:
        Xn=x2
        Yn=m15
    elif m==-20:
        Xn=x2
        Yn=m20
    elif m==-30:
        Xn=x2
        Yn=m30
    elif m==-40: #Numero asignado para nil.
        Xn=x2
        Yn=mnil
    
    #Interpolación Lineal
    term = lambda x, i: (Yn[i+1] - Yn[i])/(Xn[i+1] - Xn[i])*(x - Xn[i]) + Yn[i]
    
    #Definiendo intervalo
    for i in xrange(len(Xn)):
        if x<= Xn[i]:
             break
    y=term(x,i-1)
    
    return y
    
    
print CoolingFunction(5.5,-5) #T,Metalicidad
```

**4**. Finally, write a routine called **CoolingFunctionMetallicity** that, given any metallicity value and any temperature value, calculates the respective Cooling function.

```python
def CoolingFunctionMetallicity(temp,met):
    g=[] #Lista para guardar los valores de la primera interpolación.
    for y in xrange(len(ms)):
        li1=CoolingFunction(temp,ms[y]) #Primera Interpolación, se usa la lista de metalicidades de la celda anterior.
        g.append(li1)
    
    #Segunda interpolación usando los valores de metalicidad y de la rutina anterior.
    li2 = lambda met, i: (g[i+1] - g[i])/(ms[i+1] - ms[i])*(met - ms[i]) + g[i]
    
    for i in xrange(len(g)):
        if met<=ms[i]:
             break
    ans=li2(met,i-1)
    return ans

print CoolingFunctionMetallicity(5,-40) #Temperatura,Metalicidad
```
