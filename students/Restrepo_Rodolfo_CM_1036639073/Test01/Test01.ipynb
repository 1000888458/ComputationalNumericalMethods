#### Activity # 2 Cooling Function

```python
#Linear Interpolating Function
def LinearInterpolator( x, Xn, Yn ):
    
    #Sorting data, in case they are not
    Yn = Yn[np.argsort(Xn)]
    Xn = Xn[np.argsort(Xn)]
    
    #Detecting size of x
    try:
        Ninter = len(x)
    except:
        Ninter = 1
        x = np.array([x,])
        
    #Constructing function for each iteration
    term = lambda x, i: (Yn[i+1] - Yn[i])/(Xn[i+1] - Xn[i])*(x - Xn[i]) + Yn[i]
    
    #Detecting intervals for each x, [x_i, x_i+1] and interpolating
    y = []
    for n in xrange(Ninter):
        for i in xrange(len(Xn)):
            if x[n] <= Xn[i]:
                break
        y.append( term(x[n],i-1) )
    
    return np.array(y)
```

```python
#Cooling Funtion
def CoolingFunction(t,m):
    T = m[:,0]
    L = m[:,5]
    CF = LinearInterpolator(t,T,L)
    return CF
```

#### Activity # 3 Temperature Function

```python
#Equation for the temperature
def ET(T,M,CFo):
    ET = CoolingFunction(T,M) - CFo
    
    return float(ET)
```

```python
#Defining Bisection function
def Bisection( f, a, b, Nmax, m, CF, printer=False ):
    #verifying the STEP1, a and b with different signs
    if f(a,m,CF)*f(b,m,CF)>0:
        print "Error, f(a) and f(b) should have opposite signs"
        return False
    #Assigning the current extreme values, STEP2
    ai = a
    bi = b
    #Iterations
    n = 1
    while n<=Nmax:
        #Bisection, STEP3
        pi = (ai+bi)/2.0
        #Evaluating function in pi, STEP4 and STEP5
        if printer:
            print "Value for %d iterations:"%(n),pi
        #Condition A
        if f(pi,m,CF)*f(ai,m,CF)>0:
            ai = pi
        #Condition B
        elif f(pi,m,CF)*f(ai,m,CF)<0:
            bi = pi
        #Condition C: repeat the cycle
        n+=1
    #Final result
    return pi
```

```python
#Temperature Funtion
def Temperature(CFo,m,Ta,Tb):
    T = Bisection(ET,Ta,Tb,30,m,CFo)
    
    return T
```

```python
#Example

T = Temperature(-21.5,m5,4.0,4.5)

print T
```

#### Activity # 4 Metallicity Funtions 

```python
#Find Cooling Funtions for Temperature
def FindCFTemp(T):
    Y = [ ]
    for i in range(len(T)):
        Y.append(CoolingFunction(T[i],m5 ))
        Y.append(CoolingFunction(T[i],m00 ))
        Y.append(CoolingFunction(T[i],m05 ))
        Y.append(CoolingFunction(T[i],m10 ))
        Y.append(CoolingFunction(T[i],m15 ))
        Y.append(CoolingFunction(T[i],m20 ))
        Y.append(CoolingFunction(T[i],m30 ))
    Y = np.array(Y)
    
    return Y
```

```python
Metallicities = np.array([0.5,0.0,-0.5,-1.0,-1.5,-2.0,-3.0])
Temps=np.arange(4.5,8.5,0.5)
CFofTemps=FindCFTemp(Temps)
```

```python
plt.figure(figsize(13,6.5))
plt.plot(Metallicities,CFofTemps[0:7],label='$log(T)=4.5$',linewidth=2)
plt.plot(Metallicities,CFofTemps[7:14],label='$log(T)=5.0$',linewidth=2)
plt.plot(Metallicities,CFofTemps[14:21],label='$log(T)=5.5$',linewidth=2)
plt.plot(Metallicities,CFofTemps[21:28],label='$log(T)=6.0$',linewidth=2)
plt.plot(Metallicities,CFofTemps[28:35],label='$log(T)=6.5$',linewidth=2)
plt.plot(Metallicities,CFofTemps[35:42],label='$log(T)=7.0$',linewidth=2)
plt.plot(Metallicities,CFofTemps[42:49],label='$log(T)=7.5$',linewidth=2)
plt.plot(Metallicities,CFofTemps[49:56],label='$log(T)=8.0$',linewidth=2)

plt.legend(loc=2)
plt.title("Cooling functions for different temperatures", fontsize = 15)
plt.xlabel("Metallicity",fontsize = 15)
plt.ylabel("$Log_{10} [\Lambda(T)] [ergs\, cm^3 s^{-1}]$", fontsize = 15)
plt.grid()
```
