**2.** Routine **CoolingFunction** that, given any logarithmic temperature value $\log10(T)$ and a metallicity and using the provided data, interpolates (using by Linear interpolation) the value of the respective cooling function value.

```python
#Linear Interpolating Function
def LinearInterpolator( x, Xn, Yn ):  
    #Sorting data, in case they are not
    Yn = Yn[np.argsort(Xn)]
    Xn = Xn[np.argsort(Xn)]
    #Detecting size of x
    try:
        Ninter = len(x)
    except:
        Ninter = 1
        x = np.array([x,])      
    #Constructing function for each iteration
    term = lambda x, i: (Yn[i+1] - Yn[i])/(Xn[i+1] - Xn[i])*(x - Xn[i]) + Yn[i]
    #Detecting intervals for each x, [x_i, x_i+1] and interpolating
    y = []
    for n in xrange(Ninter):
        for i in xrange(len(Xn)):
            if x[n] <= Xn[i]:
                break
        y.append( term(x[n],i-1) )
    return np.array(y)
```

```python
def CoolingFunction(M):
    dat=np.loadtxt('m'+data[M]+'.dat')
    Xn=(dat[:,0])
    Yn=(dat[:,5])
    #Obtaining linear interpolation
    for T in Xn:
        x=T
        y = LinearInterpolator( x, Xn, Yn )
        #Evaluando en los valores dados
        print 'El valor de la funcion de enfriamiento para T=%.2f y M=%s es: %.2f' %(T,data[M],y)


    #Obtaining linear interpolation plotting with metalicity fixed
    plt.figure( figsize=(12,6) )
    Ninter = 100
    x = np.linspace(4, 8.5, Ninter )
    y = LinearInterpolator( x, Xn, Yn )
    #Plotting
    plt.plot( x, y, color="green", linewidth=2, label="linear interpolation" )
    plt.plot( Xn, Yn, "o", color="red", label='data('+'m='+data[M]+')')
    #Formatting
    plt.legend()
    plt.grid()
    plt.title('Linear interpolation with M fixed', fontsize=16)
    plt.xlabel('$\log_{10}(T)$ $[K]$', fontsize=15)
    plt.ylabel('$\log_{10}[\Lambda(T)]$ $[erg cm^3 s^{-1}]$', fontsize=15)
```

**4. Metallicity functions**
A very interesting application of interpolation techniques is the estimation of Cooling functions in terms of metallicity. For this we need to perform a double interpolation procedure.

Given a fixed value of temperature $T_0$, you can evaluate all the Cooling functions for different metallicities, obtaining a set of values $\{m_i\}_i$ for the metallicities and $\{\Lambda_i(T_0)\}_i$ for the respective Cooling functions. This can be tought as tracing a vertical line in the previous figure passing through the given value $T_0$, and then taking all the intercepted values in the Cooling functions.

```python
def InterceptCoolingfunction(T):
    for M in data:
        dat=np.loadtxt('m'+M+'.dat')
        Xn=(dat[:,0])
        Yn=(dat[:,5])
        #Obtaining linear interpolation
        Ninter = 100
        x=T
        y = LinearInterpolator( x, Xn, Yn )
        #Evaluando en los valores dados
        print 'El valor de la funcion de enfriamiento para T=%.1f y M=%s es: %.2f' %(T,M,y)
```

En particular se ha encontrado los interceptos con las temperaturas $T={4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0}$ para cada una de las metalicidades (como se muestra en la figura inferior), sin embargo el usuario puede decidir entrar un valor $T_0$ en particular dentro del intervalo [4,8.5]*
