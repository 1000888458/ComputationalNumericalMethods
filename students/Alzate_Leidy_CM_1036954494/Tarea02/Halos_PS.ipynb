## Press-Schechter

The formation of large scale structures have been obtain over the histoy through simulations. The Press-Schechter formalism is a model that gives an approximation of the number of objects that contain a given volume of the Universe with certain range of mass.

For this task has been considered the approximation:

$N(M)=1.018\times10^{14}M^{-0.889}\times \exp(-9.7\times10^{32}M^{0.222})$


```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Reading external data
data=np.loadtxt('halos.dat')
mhalo=data[:,3]
X=np.log10(mhalo)

#Prediction of the Press-Schechter formalism
Nm=1.018e14*mhalo**(-0.889)*np.exp(-9.7e-32*mhalo**0.222)
```

```python
#DATA PLOTTING
plt.hist(X,bins=200, histtype='stepfilled', color='#4169e1', label="Histogram")
plt.semilogy(X, Nm, color="#6b8e23", linewidth=1.7, label="Press-Schechter")

#GRAPHICS ADJUSTMENT
#Legend
plt.legend(loc='upper right')
#Grid
plt.grid(True)
#Range
plt.xlim(10.9,15.1)
plt.ylim(1,2e4)
#Axis labels
plt.title('Press-Shechter approximation')
plt.xlabel('$log(mass)$')
plt.ylabel('$log(N(M))$')
```

From above figure we can observe that the resolution of the halos of the simulation is close to $10^11$ orders of magnitude, furthermore it is possible to note that the simulation satisfies the Press-Schetter approximation only up masses of $10^13$ orders of magnitude.

```python
#Plotting the catalogue
fig=plt.figure( figsize = (10,10) )
ax=fig.add_subplot(111, projection='3d')
x=data[:,0]
y=data[:,1]
z=data[:,2]

#More massive halos (>1e13)
M=data[data[:,3]>1e13]
X=M[:,0]
Y=M[:,1]
Z=M[:,2]

#Plotting massive halos
ax.scatter(x,y,z, color='black', s=0.01, marker='.')
ax.scatter(X,Y,Z, color='red', s=1, marker='o')
ax.set_xlim(0, 250)
ax.set_ylim(0, 250)
ax.set_zlim(0, 250)
```

Distribution of halos represented in the previous scatter, shows that they are more likely to be in the range of coordinates between 100 and 200 in $X$, $Y$ and $Z$ axes.
