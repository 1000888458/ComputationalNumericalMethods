# Root finding & Interpolation summary
* we know how the algorithms work 
  * Bisection
  * Newton Rapson
* in real situations we use:
  * scipy.optimize.brentq (kind of bisection)
  * scipy.optimize.newton (Newton-Rapson)

```python
import scipy.optimize as optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f_cost(x):
    return x**2 - 6
```

```python
x=np.linspace(-5,5,100)
fig,ax=plt.subplots(1,1)
ax.plot(x,f_cost(x))
```

```python
root=optimize.brentq(f_cost,-4,0)
print(root)
```

```python
root=optimize.brentq(f_cost,0,4)
print(root)
```

```python
root=optimize.newton(f_cost,-4)
print(root)
```

```python
def f_cost(x,c):
    return x**2 - c
```

```python
x=np.linspace(-5,5,100)
fig,ax=plt.subplots(1,1)
ax.plot(x,f_cost(x,6))
```

```python
c=6
optimize.brentq(f_cost,-4,0,args=(c,))
```

```python
c=6
optimize.newton(f_cost,-4,args=(c,))
```

```python
x=np.linspace(-5,5,100)
fig,ax=plt.subplots(1,1)
ax.plot(x,f_cost(x,-1))
```

```python
c=-1
optimize.newton(f_cost,-4,args=(c,))
```

```python
c=-1
try:
    root=optimize.newton(f_cost,-4,args=(c,))
except:
    root=np.nan
```

```python
root
```

```python
def my_root_newton(f_cost,start,c):
    try:
        root=optimize.newton(f_cost,start,args=(c,))
    except:
        root=np.nan
    return root    
```

```python
c=-2
start=-100
root=my_root_newton(f_cost,start,c)
print(root)
```

```python
def f_cost(x,c):
    return x**2 - c
```

```python
c_vec=np.linspace(-20,20,10)

results=[]
start=10

for c in c_vec:
    root=my_root_newton(f_cost,start,c)
    results.append(root)
```

```python
def my_interp_spline(x,y,xinterp):
    idx=~np.isnan(y)
    x=x[idx]
    y=y[idx]
    f_spline=interp1d(x,y,kind='cubic')
    yinterp=f_spline(xinterp)
    return yinterp
```

```python
def my_interp_spline(x,y,xinterp,kind):
    idx=~np.isnan(y)
    x=x[idx]
    y=y[idx]
    f_spline=interp1d(x,y,kind=kind)
    yinterp=f_spline(xinterp)
    return yinterp
```

```python
yinterp=my_interp_spline(x,y,xinterp,'cubic')
```
