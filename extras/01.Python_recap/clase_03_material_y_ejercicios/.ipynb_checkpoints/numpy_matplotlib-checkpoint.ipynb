```python
import numpy as np
```

```python
x = np.linspace(-np.pi,np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.exp(-x*0.5)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,5))
ax.plot(x,y1,'--r',label='sin(x)')
ax.plot(x,y2,'-b',label='cos(x)')
ax.plot(x,y3,'.g',label='custom')
ax.legend()
ax.set_title('my funcions')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
fig.savefig('my_plot.png')
```

```python
x=np.linspace(-np.pi,np.pi,100)
y4=[]
for i in range(len(x)):
    y4.append(np.sin(x[i]))

fig,ax=plt.subplots(1,1)
ax.plot(x,y4)
```

# Plotting distributions

```python
# normal distribution
#n_dist=np.random.normal(mu,sigma,numbers)
normal_dist=np.random.normal(160,5,1000000)
print(type(n_dist))
fig,ax = plt.subplots(1,1)
bins=40
ax.hist(normal_dist,bins)
print()
```

# uniform distribution

```python
unif_distr=np.random.uniform(0,10,1000000)
fig,ax=plt.subplots(1,1)
bins=40
ax.hist(unif_distr,bins)
print()
```
