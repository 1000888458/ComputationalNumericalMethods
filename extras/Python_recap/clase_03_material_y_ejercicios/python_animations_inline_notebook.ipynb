```python
%matplotlib notebook

import numpy as np
from matplotlib import pyplot as plt
from time import sleep

x=np.linspace(-2*np.pi,2*np.pi,100)
fig,ax=plt.subplots(1,1)

t=0.0
while True:
    ax.cla()
    ax.plot(x,np.cos(x-t),'-k')
    t=t+0.1
    fig.canvas.draw()
#    sleep(0.0)
```

```python
%matplotlib notebook

import numpy as np
from matplotlib import pyplot as plt
from time import sleep

time_lap=[0]
value=[0]
t=1.0
fig,ax=plt.subplots(1,1)

while True:
    ax.cla()
    ax.plot(time_lap,value,'-k')
    fig.canvas.draw()
    sleep(t)
    time_lap.append(time_lap[-1]+t)
    value.append(np.random.uniform(0,1))
```
